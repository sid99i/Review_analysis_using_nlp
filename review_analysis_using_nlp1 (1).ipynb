```python
import pandas as pd
```

```python
#import the dataset as a csv file
dataset = pd.read_csv('/content/Restaurant_Reviews-211105-185708.tsv', delimiter='\t',quoting=3)
```

```python
dataset
```

```python
#importing the nltk and re library for preprocessing and cleaning of data
import nltk
import re
```

```python
#updating the stopwords in the nltk library
nltk.download('stopwords')
```

```python

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

corpus = []
ps = PorterStemmer()
all_stopwords = stopwords.words('english')
all_stopwords.remove('not')

for line in range(0,1000):
  review = re.sub('[^a-zA-Z] ', ' ', dataset['Review'][line])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1500)
x = cv.fit_transform(corpus).toarray()
y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
y_test
```

```python
x_test
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train, y_train)
y_pred = gnb.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
accuracy_score(y_test,y_pred)
```
